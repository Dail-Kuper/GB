```python
import requests
import pprint
from lxml import html
from lxml import etree
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0'}
url = 'https://russian.rt.com/'

answer = requests.get(url=url, headers=header)
```

```python
print(answer)
print(answer.status_code)
print(answer.ok)
```

```python
news = []
mydict = {}

sublist = []
news_name = ""
news_ref = ""
news_time = ""
news_url = ""
mystr = ""
i = 1

page = html.fromstring(answer.text)

while  i < 4:
    
    news_name = "//div[@class='listing__card listing__card_main-promobox listing__card_main-promobox_%d']/div/div[3]/a/text()" % (i)
    news_ref = "//div[@class='listing__card listing__card_main-promobox listing__card_main-promobox_%d']/div/div[3]/a/@href" % (i)
    news_time = "//div[@class='listing__card listing__card_main-promobox listing__card_main-promobox_%d']/div/div[4]/div[1]/text()" % (i)
    
    news_name = page.xpath(news_name)
    news_ref = page.xpath(news_ref)
    news_time = page.xpath(news_time)
    
    mystr = news_name[0]
    news_name = mystr[9:-7]
    
    mystr = news_time[0]
    news_time = ""
    for j in range(len(mystr)):
        if mystr[j] in "0123456789:":
            news_time = news_time + mystr[j]
    
    mydict = {}
    mydict["news_name"] = news_name
    mydict["news_url"] = url[:-1] + news_ref[0]
    mydict["news_time"] = news_time
    mydict["news_sourse"] = "RT"
    
    news.append(mydict) 
    
    i = i +1
    
pprint.pprint(news)

```
