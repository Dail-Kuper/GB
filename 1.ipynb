```python
param = {
'limit': '222',
'offset': '0',
'latitude': '47.265322',
'longitude': '39.72175',
'fastFilters': 'group',
'cityId': '10'
}
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0'}

url = 'https://api.delivery-club.ru/api1.3/screen/web/components'
req = requests.get(url=url, params=param, headers=header)
# limit=48&offset=0&latitude=47.265322&longitude=39.72175&fastFilters=group&cityId=10
# https://api.delivery-club.ru/api1.3/screen/web/components?limit=48&offset=0&latitude=47.265322&longitude=39.72175&fastFilters=group&cityId=10&cacheBreaker=1663926807
```

```python
mylist1 = []
for component in components:
    print(component["vendor"]["name"], component["vendor"]["cuisines"])
    cuisines = component["vendor"]["cuisines"]
    for one in cuisines:
       mylist1.append(one)
```

```python
myset = ()
myset = set(mylist1)
mydict = {}

print("Всего видов категорий - " + str(len(myset)))
print()

for setelement in myset:
    count = 0
    #print(setelement)
    for i in range(len(mylist1)):
        if setelement == mylist1[i]:
            count +=1
           
    print(setelement + " - " + str(count))
    mydict[setelement] = count
```

```python
list_cuisines = []
list_count = []

mydict = dict(sorted(mydict.items(), key=lambda item: item[1]))
print(mydict)
for key in mydict:
    list_cuisines.append(key)
    list_count.append(mydict[key])
    
list_cuisines.reverse()
list_count.reverse()
```

```python
import matplotlib.pyplot as plt
from decimal import Decimal

%config inlinebackend.figure_format = 'svg'

fig, graph  = plt.subplots()
graph.bar(list_cuisines, list_count)


plt.ylabel("Частота повторений", fontsize = 20)
plt.xlabel("Теги", fontsize = 20)
plt.title("График частоты категорий для первых 200 заведений, представленных в Ростове-на-Дону", fontsize = 22, fontweight = 'bold')

plt.yticks(fontsize = 18)
plt.xticks(fontsize = 18, rotation = 90)

fig.set_figwidth(20)
fig.set_figheight(20)
print(sum(list_count))

for i in range(len(list_count)):
    number = Decimal(list_count[i]/sum(list_count) * 100)
    plt.annotate(str(number.quantize(Decimal("1."))) + " %", xy=(list_cuisines[i],list_count[i]), ha='center', va='bottom', fontsize = 16)

plt.show()
```

```python
mylist1 = []
mylist2 = []
mysublist = []
mystr = ""
hours_to_min = 0
found_min = False
found_digit = False
a = ""
b = 0

for key in mydict:
    mylist1.append(mydict[key]["time"])

for i in range(len(mylist1)):
    mystr = mylist1[i]
    mysublist = mystr.split()
    #print(mylist1[i])
    
    hours_to_min = 0
    found_min = False
    found_digit = False
    a = ""
    b = 0
    
    for j in range(len(mysublist)):
        if "час" in mysublist[j]:
            hours_to_min = 60 * int(mysublist[j-1])
        if "завтра" in mysublist[j]:
            hours_to_min = 60 * 24
        
        if "мин" in mysublist[j]:
            found_min = True
            mystr = mysublist[j-1]
            for k in range(len(mystr)):
                if mystr[k] in "0123456789":
                    found_digit = True
                    a = a + mystr[k]
                elif mystr[k] not in "0123456789" and found_digit == True:
                    b = int(a)
                    a = ""

                if k == len(mystr)-1 and b == 0:
                    mylist2.append(int(a) + hours_to_min)
                if k == len(mystr)-1 and b != 0:
                    mylist2.append(int(((int(a) + b)/2) + hours_to_min))
                    #print(a, b)
                        
                    
                    
            
        if found_min == False and j == len(mysublist)-1:
            mylist2.append(hours_to_min)
            #print(hours_to_min)
            
            
print(mylist2)
```
